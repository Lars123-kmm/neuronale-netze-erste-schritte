```python
# function that takes 2 numbers a input and outputs their average
def avg(x, y):
    print("first input ist", x)
    print("second input ist", y)
    a = (x + y) / 2.0
    print("average is", a)
    return a
# Funktion avg(x,y) verwendet zwei bekannte Elemente, sie gibt Zahlen aus, die die Funktion
# bei ihrem Aufruf übergehen bekommt. Die Anweisung berechnet (x+y)/2.0 und weist den Ergebniswert der Variablen a zu.
# Die nächste Anweisung gibt den Mittelwert aus , um wieder den Ablauf im Code nachvollziehbar zu machen"
# Mit return a wird Pyhton gesagt, was die Funktion zurückgeben soll.
# Beim Ausführen des Codes, passiert nicht, er keine Zahlen, denn man hat die Funktion lediglich definiert, aber noch nicht verwendet.
# Python hat diese Funktion zur Kenntnis genommen und hält sie bereit, wenn man sie nutzen will
# Eine wiederverwendbare Funktion ist definiert.
```

```python
# Library imports
import numpy
import scipy.special
# scipy.special enthält die Sigmoid-Funktion expit()
```

```python
# NN Klasse
class neuralNetwork:

    # init vom Netz
    def __init__(self, inputnodes, hiddennodes, outputnodes, learningrate):

        # Knoten speichern
        self.inodes = inputnodes      # Input-Knoten
        self.hnodes = hiddennodes     # Hidden-Knoten
        self.onodes = outputnodes     # Output-Knoten

        # Gewichte initialisieren
        # Normalverteilung, damit Startwerte nicht zu extrem sind
        self.wih = numpy.random.normal(
            0.0, pow(self.hnodes, -0.5),
            (self.hnodes, self.inodes)
        )   # Gewichte input -> hidden

        self.who = numpy.random.normal(
            0.0, pow(self.onodes, -0.5),
            (self.onodes, self.hnodes)
        )   # Gewichte hidden -> output

        # Lernrate merken
        self.lr = learningrate        # meine Lernrate

        # Aktivierungsfunktion (Sigmoid)
        # Werte zwischen 0 und 1 zu bekommen
        self.activation_function = lambda x: scipy.special.expit(x)

        pass

    # Training für später
    def train(self):
        pass

    # Netz abfragen
    def query(self, inputs_list):

        # Inputs in 2D-Form bringen (fürs Dot-Produkt)
        inputs = numpy.array(inputs_list, ndmin=2).T

        # hidden layer berechnen
        hidden_inputs = numpy.dot(self.wih, inputs)      # Signale in Hidden
        hidden_outputs = self.activation_function(hidden_inputs)   # aktivieren

        # output layer berechnen
        final_inputs = numpy.dot(self.who, hidden_outputs)          # Signale in Output
        final_outputs = self.activation_function(final_inputs)      # aktivieren

        return final_outputs    # Ergebnis ausgeben
```
