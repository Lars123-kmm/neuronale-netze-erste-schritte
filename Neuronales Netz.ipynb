```python
# function that takes 2 numbers a input and outputs their average
def avg(x, y):
    print("first input ist", x)
    print("second input ist", y)
    a = (x + y) / 2.0
    print("average is", a)
    return a
# Funktion avg(x,y) verwendet zwei bekannte Elemente, sie gibt Zahlen aus, die die Funktion
# bei ihrem Aufruf übergehen bekommt. Die Anweisung berechnet (x+y)/2.0 und weist den Ergebniswert der Variablen a zu.
# Die nächste Anweisung gibt den Mittelwert aus , um wieder den Ablauf im Code nachvollziehbar zu machen"
# Mit return a wird Pyhton gesagt, was die Funktion zurückgeben soll.
# Beim Ausführen des Codes, passiert nicht, er keine Zahlen, denn man hat die Funktion lediglich definiert, aber noch nicht verwendet.
# Python hat diese Funktion zur Kenntnis genommen und hält sie bereit, wenn man sie nutzen will
# Eine wiederverwendbare Funktion ist definiert.
```

```python
avg(2, 4)
```

```python
avg(200, 301)
```

```python
# class for a dog object
class Dog:
    # initialisation method with internal data
    def __init__(self, petname, temp):
        self.name = petname
        self.temperature = temp
    
    # get a status
    def status(self):
        print("Dog name:", self.name)
        print("Dog temperature:", self.temperature)

    # set temperature
    def set_temperature(self, temp):
        self.temperature = temp

    # dogs can bark()
    def bark(self):
        print("Woof!")
```

```python
# Create a new dog object from the Dog class
lassie = Dog("Lassie", 38.5)
lassie.status()
```
